```python
# Importing libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
# Importing the dataset from Excel. The information is stored in the object 'data'

data = pd.read_excel("Superstore.xls")
```

```python
# Obtaining the information about the data
# This also provides us information about the type of the attributes in the dataset and missing values

data.info()
```

```python
# Obtaining summary statistics with the help of describe function. 
# This function returns the count, mean, standard deviation, minimum and maximum values and the quartiles of the data.

data.describe()
```

```python
# Displaying the columns of the data set

data.columns
```

# PRODUCT - WISE ANALYSIS

```python
# This code groups the data by Product ID. It then rearranges it in descending order in terms of sales.
# This helps in identifying the top 10 products which give the most sales

data1 = data.groupby(['Product ID']).sum()
data1 = data1.loc[:,['Sales']]
data2 = data1.sort_values(['Sales'], ascending=False)
data2.head(10)
```

```python
# The code helps in identifying the product which contributes the least to sales

data2.tail(10)
```

```python
# This code groups the data by Product ID. It then rearranges it in descending order in terms of profit.
# This helps in identifying the top 10  products which give the most profits

data1 = data.groupby(['Product ID']).sum()
data1 = data1.loc[:,['Profit']]
data2 = data1.sort_values(['Profit'], ascending=False)
data2.head(10)
```

```python
# The code helps in identifying the product which contributes the least to profits

data2.tail(10)
```

# CUSTOMER - WISE ANALYSIS

```python
# This code groups the data by Customer ID. It then rearranges it in descending order in terms of profit.
# This helps in identifying the top 10 customers which give the most profits

customers = data['Customer ID'].unique()
dataCustomer = data.groupby("Customer ID").sum()
dataCustomer2 = dataCustomer.sort_values(['Profit'], ascending=False)
dataCustomer2.head(10)
```

```python
# Plotting the above customer data in a bar graph (Most profitable customers)

dataCustomer2['Profit'].head(10).plot(kind='bar')
```

```python
# This code groups the data by Customer ID. It then rearranges it in descending order in terms of profit.
# This helps in identifying the 10 customers which give the least profits

customers = data['Customer ID'].unique()
dataCustomer = data.groupby("Customer ID").sum()
dataCustomer2 = dataCustomer.sort_values(['Profit'], ascending=False)
dataCustomer2.tail(10)
```

```python
# Plotting the above customer data in a bar graph (Least profitable customers)

dataCustomer2['Profit'].tail(10).plot(kind='bar')
```

```python
# This code gives a sales vs profits bar graph for most profitable customers

dataCustomer3 = dataCustomer2[["Profit","Sales"]].head(10)
dataCustomer3.plot(kind="bar")
```

```python
# This code gives a sales vs profits bar graph for least profitable customers

dataCustomer3 = dataCustomer2[["Profit","Sales"]].tail(10)
dataCustomer3.plot(kind="bar")
```

```python
# This code gives an overall Sales vs Profits

data[["Sales","Profit"]].plot.area(stacked=False,title="Sales vs. Profit")
```

```python
# This code gives a yearly data of sales vs profits over the years as a trend line in line chart

matplotlibdf = data[['Order Date', 'Sales', 'Profit']].copy()
matplotlibdf['Order Date'] = pd.to_datetime(matplotlibdf['Order Date'], format='%d/%m/%y')
linechartData = matplotlibdf.groupby([pd.Grouper(freq='MS',key='Order Date')]).sum()
linechartData.plot(figsize=(20,10))
```

# SEGMENT WISE ANALYSIS

```python
# This code gives an aggregate of sales and profits by grouping it on the basis of Segment

segment = data\
.groupby(['Segment'])\
.agg({'Sales': 'sum', 'Profit': 'sum'})\
.reset_index()\
.sort_values('Sales', ascending=False)

segment
```

```python
# This code plots a bar chart of segment wise sales with segment on the y-axis and sales on the x-a

f, ax = plt.subplots(figsize=(10, 6))
viz = sns.barplot(
    data=segment,
    x='Sales',
    y='Segment'
)
```

# CATEGORY - WISE ANALYSIS

```python
# This code gives an aggregate of sales and profits by grouping it on the basis of Category

cat = data\
.groupby(['Category'])\
.agg({'Sales': 'sum', 'Profit': 'sum'})\
.reset_index()\
.sort_values('Sales', ascending=False)

cat
```

```python
# This code plots a bar chart of category wise sales with category on the y-axis and sales on the x-axis
# We can see that Technology is the highest in Sales, followed by Furniture and then Office Supplies

f, ax = plt.subplots(figsize=(10, 6))
viz = sns.barplot(
    data=cat,
    x='Sales',
    y='Category'
)
```

```python
# This code gives a Sales vs Profit analysis based on the Category
# Index : 2-Technology, 1-Office Supplies, 0-Furniture

datacat = cat[["Profit","Sales"]]
datacat.plot(kind="bar")
```

# SUB-CATEGORY WISE ANALYSIS

```python
# This code gives an aggregate of sales and profits by grouping it on the basis of SubCategory

subcat = data\
.groupby(['Sub-Category'])\
.agg({'Sales': 'sum', 'Profit': 'sum'})\
.reset_index()\
.sort_values('Sales', ascending=False)

subcat
```

```python
# This plot shows the Sales of sub-categories of products. 
# Phones and chairs have the highest sales whereas fasteners have the least sales
 
f, ax = plt.subplots(figsize=(10, 6))
viz = sns.barplot(
    data=subcat,
    x='Sales', 
    y = 'Sub-Category'
)
```

```python
# This plot shows the profits of sub-categories of products. 
# Tables, bookcases and supplies are the only sub-categories that have losses
# Copiers, phones and paper have the highest profits

f, ax = plt.subplots(figsize=(10, 6))
viz = sns.barplot(
    data=subcat,
    x='Profit', 
    y = 'Sub-Category'
)
```

# GEOGRAPHICAL ANALYSIS

```python
# This code gives an aggregate of sales and profits by grouping it on the basis of Region

reg = data\
.groupby(['Region'])\
.agg({'Sales': 'sum', 'Profit': 'sum'})\
.reset_index()\
.sort_values('Sales', ascending=False)

reg
```

```python
# This plot shows the region-wise sales  
# West has the highest sales, followed by East, Central and South respectively

f, ax = plt.subplots(figsize=(10, 6))
viz = sns.barplot(
    data=reg,
    x='Sales' ,
    y='Region'
)
```

```python
# This plot shows the region-wise profits  
# West has the highest profits, followed by East, South and Central respectively

f, ax = plt.subplots(figsize=(10, 6))
viz = sns.barplot(
    data=reg,
    x='Profit' ,
    y='Region'
)
```

```python
# This shows the state-wise sales and profits
# California has the highest sales, followed by New York 

state = data\
.groupby(['State'])\
.agg({'Sales': 'sum', 'Profit': 'sum'})\
.reset_index()\
.sort_values('Sales', ascending=False)

state
```

```python
# This plot shows the sales of states 

f, ax = plt.subplots(figsize=(25, 6))
viz = sns.barplot(
    data=state,
    x='State' ,
    y='Sales'
)
plt.xticks(rotation=-90)
```

```python
# This shows the top 10 states with highest sales

data1 = data.groupby(['State']).sum()
data1 = data1.loc[:,['Sales']]
data2 = data1.sort_values(['Sales'], ascending=False)
data2.head(10)
```

```python
# This shows the 10 states with least sales

data1 = data.groupby(['State']).sum()
data1 = data1.loc[:,['Sales']]
data2 = data1.sort_values(['Sales'], ascending=False)
data2.tail(10)
```

```python
# This shows the top 10 states with highest profits

data1 = data.groupby(['State']).sum()
data1 = data1.loc[:,['Profit']]
data2 = data1.sort_values(['Profit'], ascending=False)
data2.head(10)
```

```python
# This shows the 10 states with least profits

data1 = data.groupby(['State']).sum()
data1 = data1.loc[:,['Profit']]
data2 = data1.sort_values(['Profit'], ascending=False)
data2.tail(10)
```

```python
# This plot shows the profit/loss of states 
# There are about 10 loss-making states  
# Arizona, Colorado, Florida, Illinois, North Carolina, Ohio, Oregon, Pennsylvania, Tennessee and Texas

f, ax = plt.subplots(figsize=(25, 6))
viz = sns.barplot(
    data=state,
    x='State' ,
    y='Profit'
)
plt.xticks(rotation=-90)
```
